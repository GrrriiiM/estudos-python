```python
f = Lambda(var("x"), x**2 + 20*x - 3)
f
```

```python
f1 = Lambda(x, diff(f(x), x))
f1
```

```python
f2 = Lambda(x, integrate(f(x), x))
f2
```

```python
p = plot(f(x), line_color='blue', show=False)
p.extend(plot(f1(x), line_color='red', show=False))
p.extend(plot(f2(x), line_color='green', show=False))
p.show()
```
